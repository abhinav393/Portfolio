# Recursive function

Recursion is a method of programming or coding a problem, in which a function calls itself one or more times in its body.Lets take an example of Fibonacci series,the first few terms of Fibonacci series are:

![title](fs.png)

- The 1st two terms are one’s, there after each terms is the some of the previous two terms. For example:-
    - 1+1=2, 2 is the 3rd term from there 
    - 1+2 =3, 3 is the 4th term from there
    - 2+3=5, 5 is the 5th term so on and so forth…


- Our goal is to write a function to return nth term of Fibonacci sequence. Our function should be:-
  - Fast
  - Clearly written 
  - Rock solid.


- You may think , this is simple but by the time we are done with our function, it will be more complex than you may expect.
- By the way some people argue that the sequence should start with 0 & 1, if you encounter this situation don’t panic.Other than the zero the two sequence are identical. People love to debate, on smallest of details.


![title](fs1.png)

###### Let’s now write the function, we begin by giving our function a name ‘fibo’, there will a single input ‘n’,this function will return the nth term,

```python
#fibonacci series by recurssion
def fibo(n):
    if n==1:
        return 1
    elif n==2:
        return 1
    elif n>2:
        return fibo(n-1) + fibo(n-2)

for i in range(1,11):
    print(i,":- ", fibo(i))
```

So our Fibonacci function worked, but does this mean we are done?
Before we move on lets test the function for few more value. What if we display the first 100 terms of the fibbonacci sequence.


```python
for i in range(1,101):
    print(i,":- ", fibo(i))
```

After the 29th term of the series you will realize that your function is slowing down and that’s disappointing. Our function is virtually useless after few dozen of terms.Why is this? And how can we fix this?

Note: that was not an error, just because it got stuck at 33rd term for so long, i interrupted the kernal.

To see why the system is slow look what happens when the system computes the 5th term ,Because this is a recursive function,

![title](fs3.png)

- So this way our computer is repeating it self needlessly.Since Fibonacci(1) and Fibonacci(2) are both 1, we finally get our answer 5.Our simple recursive function is careless, we are make it to repeat it self over and over and over again.

- There is a simple cure to this problem i.e Memoization.the idea behind memorization is simple store the value of the recent function call, so the future call do not need to repeat itself.


###### There are several ways to use memorization in python.
- Implementing memorization explicitly.
- Using built-in python tool


To begin we should create a dictionary  ‘cache’  which will store the recent function calls,then write the Fibonacci function to make use of cached value.We do not want to simply return the value, instead we will compute the value , cache it and then return it.

```python
cache={}
def fibo(n):
    if n in cache:
        return cache[n]
    if n ==1:
        value =1
    elif n==2:
        value =1
    elif n>2:
        value = fibo(n-1) + fibo(n-2)
    
    cache[n] = value
    return value

for i in range(1,101):
    print(i,":- ",fibo(i))
```

So now you can use the code and checkout yourself, it works perfectly fine. Now that you have seen the explicit use of memoization, but the simpler approach would be using the built-in lru_cache module, stands for Least Recently Used Cache.

To use this,import the module and then  type @lru_cache(maxsize=1000) before our function. In parentheses, enter the number of value to cache by default python will cache 128 most recent values. So if you do not specify the maxsize you will still see the benefits.

```python
from functools import lru_cache
@lru_cache(maxsize =1000)

def fibo(n):
    if n==1:
        return 1
    elif n==2:
        return 2
    elif n>2:
        return fibo(n-1) + fibo(n-2)
    
for i in range(1,100):
    print(i,":- ", fibo(i))
```

```python
#Here’s, the example of Fibonacci series without recursion.

def fibo(n):
    a,b = 0,1
    for i in range(n):
        a=b
        b=a+b
        print(a,"\n")
    return b
```

```python
fibo(100)
```
