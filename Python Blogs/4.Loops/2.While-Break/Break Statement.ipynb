#### Infinity loop

```python
# Program on infinity loop
#Do not run this code SINCE IT WONT STOP
#IN JUPYTER NOTEBOOK PRESS THAT SQUARE BUTTON WHICH INTERUPT THE KERNAL
#IF YOU ARE USNING PYTHON IDE, THEN
# use ctrl + c to terminate this infinite loop.
while True:
    print('hello world')
```

- Since the condition True, will always be True this loop will keeps looping forever. At certain situation Press ctrl + c for keyboard interrupt error. It will crash the program & stops immediately.
- There are two new kinds of statement that can go inside of a while loop .i.e  Break Statement & Continue  Statement


```python
# Program on infinity loop
#Do not run this code in jupyter use python ide and
# use ctrl + c to terminate this infinite loop.

name=''
while True:
    print("Please type ur name.")
    name=input()
    if name == 'mrityunjay':
        print(" Thankyou ")
```

- Just because this condition will always be True so , no matter what you type either True or False , it will keep looping weather the condition is True or False .

- But, it can create an error in your program , so what will be the solution ???? 
- Well the solution is very simple a " break Statement ".


### **Break Statement:**
The break statement causes the execution to immediately jumps out of the loop, it  doesn't even checks the condition again , it just immediately jumps out to execute the next line after of the loop.


```python
name= ''
while True:
    print("Please type ur name.")
    name=input()
    if name == 'mrityunjay':
        break
print(" Thankyou ")
```

![title](8.png)

The break statement causes the executions to jumps out of the loop and then continue on with the rest of the program & prints Thank you
